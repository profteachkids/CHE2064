```
import torch
import torch.nn as nn
import torchvision as tv
from plotly.subplots import make_subplots
import sklearn
```

```
x = torch.linspace(0,10, 25).unsqueeze(0)
y= torch.sin(x)
x_norm = (x-x.mean())/x.std()
```

```
lr=1e-1

w1=torch.distributions.Normal(0, 1/20**0.5).sample((20,1)).requires_grad_()
b1=torch.zeros((20,1),requires_grad=True)

for i in range(20000):
  o1=torch.tanh(w1@x_norm+b1)
  o2=torch.tanh(w2@o1+b2)
  predict=torch.sum(o2,dim=0)
  loss = torch.mean((predict-y)**2)
  if i % 1000 ==0:
    print(i, loss)
  loss.backward()
  with torch.no_grad():
    w1.sub_(lr * w1.grad)
    b1.sub_(lr * b1.grad)
    w2.sub_(lr * w2.grad)
    b2.sub_(lr * b2.grad)
    w1.grad=None
    b1.grad=None
    w2.grad=None
    b2.grad=None


```

```
fig= make_subplots()
fig.add_scatter(x=x.squeeze(),y=y.squeeze(), mode='markers')
fig.add_scatter(x=x.squeeze(), y=predict.detach().numpy(), mode='lines')
fig.update_layout(width=800,height=500)
fig.show()
```

```
x = torch.linspace(0,10, 25).unsqueeze(1)
y= torch.sin(x)

x_norm = (x-x.mean())/x.std()

model = nn.Sequential(nn.Linear(1,20),
                      nn.Tanh(),
                      nn.Linear(20,20),
                      nn.Tanh(),
                      nn.Linear(20,1))

loss_fn = nn.MSELoss()
optim = torch.optim.Adam(model.parameters())
for i in range(10000):
  loss = loss_fn(model(x_norm), y)
  if i % 1000 == 0:
    print(i, loss)
  optim.zero_grad()
  loss.backward()
  optim.step()
```

```
fig= make_subplots()
fig.add_scatter(x=x.squeeze(),y=y.squeeze(),mode='markers')
fig.add_scatter(x=x.squeeze(), y=model(x_norm).squeeze().detach().numpy(), mode='lines')
fig.update_layout(width=800,height=500)
fig.show()
```
